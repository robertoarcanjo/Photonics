Considering the problem: [Partial Diferential Equations - Example](../Notes/PDE_Example.pdf)



```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

def U_v1(Vo,x,y,n,a,b): 
    U = 0

    for i in range(n):
        U += (4*Vo/((2*i+1)*np.pi))*(1/np.sinh((2*i+1)*np.pi*b/a))*np.sin((2*i+1)*np.pi*x/a)*np.sinh((2*i+1)*np.pi*(b-y)/a)
    
    return U

def U_v2(Vo,x,y,n,a,b):
    U = 0

    for i in range(n):
        U += (4*Vo/((2*i+1)*np.pi))*(1/np.sinh((2*i+1)*np.pi*a/b))*np.sinh((2*i+1)*np.pi*x/b)*np.sin((2*i+1)*np.pi*y/b)
    
    return U

def U_v3(Vo,x,y,n,a,b): 
    U = 0

    for i in range(n):
        U += (4*Vo/((2*i+1)*np.pi))*(1/np.sinh((2*i+1)*np.pi*b/a))*np.sinh((2*i+1)*np.pi*y/a)*np.sin((2*i+1)*np.pi*x/a)
    
    return U

def U_v4(Vo,x,y,n,a,b): 
    U = 0

    for i in range(n):
         U += (4*Vo/((2*i+1)*np.pi))*(1/np.sinh((2*i+1)*np.pi*a/b))*np.sin((2*i+1)*np.pi*y/b)*np.sinh((2*i+1)*np.pi*(a-x)/b)
    
    return U

a = 5
b = 5

n = 100

V1 = 1 
V2 = 2
V3 = 3
V4 = 4

X = np.arange(0, a, 0.1)
Y = np.arange(0, b, 0.1)
X, Y = np.meshgrid(X, Y)
```

```python
Z1 = U_v1(V1, X, Y, n, a, b)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z1)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('U')

plt.show()

# Plot in 2d
fig, ax = plt.subplots()
CS = ax.contourf(X, Y, Z1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Contour plot of U1')
plt.colorbar(CS)
plt.show()
```

```python
Z2 = U_v2(V2, X, Y, n, a, b)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z2)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('U')

plt.show()

# Plot in 2d
fig, ax = plt.subplots()
CS = ax.contourf(X, Y, Z2)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Contour plot of U2')
plt.colorbar(CS)
plt.show()
```

```python
Z3 = U_v3(V3, X, Y, n, a, b)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z3)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('U')

plt.show()

# Plot in 2d
fig, ax = plt.subplots()
CS = ax.contourf(X, Y, Z3)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Contour plot of U3')
plt.colorbar(CS)
plt.show()
```
